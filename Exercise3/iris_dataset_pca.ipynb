```python
# Importing the libraries
```

```python
from sklearn import datasets
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
```

```python
# Loading the iris dataset
```

```python
irisData = datasets.load_iris()
```

```python
# Declaring the variables
```

```python
# Setting the pca components to 2 since we want to have 2 dimensions
```

```python
pca = PCA(2)
print(pca)
```

```python
# Now we want to iniliaze the x and y variables
```

```python
x, y = irisData.data, irisData.target
x_proj = pca.fit_transform(x)
print(x_proj.shape)
```

```python
# As we can see the shape of the x_proj is 150,2 which means that we have used principal component analysis to reduce the dimensions of the data from 4 to 2
```

```python
# Now we want to plot the data
```

```python
plt.scatter(x_proj[:, 0], x_proj[:, 1], c=y)
plt.title('Real Classification')
plt.show()
```

```python
# As we can see from the plot that the data is not linearly seperable
```

```python
# Now we want to use the kmeans clustering algorithm to cluster the data
```

```python
# Setting the amount of clusters to three since we have 3 different classes in the data
# We have the class iris-setosa, iris-versicolor and iris-virginica
k = 3
kmean = KMeans(n_clusters=k, random_state=0, n_init=10).fit(x_proj)
```

```python
# Now we want to plot the data
```

```python
labels = kmean.labels_
clusters = kmean.cluster_centers_
print(clusters)
```

```python
# Now we want to plot the data
```

```python
plt.subplot(1, 2, 2)
plt.scatter(x_proj[:, 0], x_proj[:, 1], c=labels)
plt.scatter(clusters[:, 0], clusters[:, 1], marker='*', s=200, c='#050505')
plt.title('K-means')

# Plot the Original Classifications
plt.subplot(1, 2, 1)
plt.scatter(x_proj[:, 0], x_proj[:, 1], c=y, s=40)
plt.title('Real Classification')
plt.show()
```

# Clustering
Shown from the diagram we can see that seperating the data into 3 clusters representing a pretty good clustering of the data.
Q: What would it look like if we used DBSCAN instead of K-means?
using dbscan could cause an issue since, as the real classificaition shows, the data points, althoug they are 2 different classes, they are placed closeley together. 

```python
db = DBSCAN(eps=0.5, min_samples=5).fit(x_proj)
labels = db.labels_
clusters = db.components_
```

```python
# Now we want to plot the data
```

```python
plt.subplot(1, 2, 2)
plt.scatter(x_proj[:, 0], x_proj[:, 1], c=labels)
plt.title('DBSCAN')
# Plot the Original Classifications
plt.subplot(1, 2, 1)
plt.scatter(x_proj[:, 0], x_proj[:, 1], c=y, s=40)
plt.title('Real Classification')
plt.show()
```

# Result
Given both the K-means and the DBSCAN we can see that the Kmeans algorithm does a better job at clustering the data, than DBSCAN does.
